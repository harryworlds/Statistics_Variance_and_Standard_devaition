```python
pet_height = [600, 470, 170, 430, 300]
pet_height.sort()
median = 0
if len(pet_height)%2 == 0:
    x = (len(pet_height))/2
    y = (len(pet_height)/2) + 1
    median = (x+y)/2
    print("n is even")
else:
    median = (len(pet_height)+1)//2
    print("n is odd")

if (len(pet_height) % 2 != 0):
    print(pet_height[median-1])
elif (len(pet_height) % 2 == 0):
    print((pet_height[math.floor(median)-1]+pet_height[math.floor(median)])/2)
```

5.Population Variance and Standard Deviation

```python
pet_height = [600, 470, 170, 430, 300]
mean = sum(pet_height)/len(pet_height)
var = sum((xi-mean)**2 for xi in pet_height)/len(pet_height)
print("Population Variance is", int(var))

# Population Standard Deviation Equation
# σ = ([Σ(x - u)^2]/N)1/2
def square_root(num):
      return num**0.5
print("Standard Deviation is",round(square_root(var),2))
```
